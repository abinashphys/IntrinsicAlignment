## $A_1$ and $\eta_1$ constant while $A_2$ and $\eta_2$ varies

## GGL

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the CSV file to inspect its content
file_path = 'Final_TATT_deltachi2.csv'
df = pd.read_csv(file_path)

# Identifying 9 unique pairs of (A1, eta1)
unique_pairs_9 = df[['A1', 'eta1']].drop_duplicates().sample(n=9, random_state=1)

# Setting up the matplotlib figure and axes grid
fig, axes = plt.subplots(3, 3, figsize=(15, 15))

# Flatten the axes array for easy iteration
axes_flat = axes.flatten()

# Plotting the data for each unique pair of A1 and eta1
for idx, (a1, eta1) in enumerate(unique_pairs_9.values):
    ax = axes_flat[idx]
    subset = df[(df['A1'] == a1) & (df['eta1'] == eta1)]
    
    # Creating a scatter plot
    points = ax.scatter(subset['A2'], subset['eta2'], c=subset['ggl: Delta chi2'], cmap='magma', marker='o', s=50)
    
    # Setting the title for each subplot
    ax.set_title(f'$A_1={a1}, \eta_1={eta1}$')

    # Removing the axis labels (they will be added back later for the outer plots)
    ax.set_xlabel('$A_2$')
    ax.set_ylabel('$\eta_2$')

    # Adding a colorbar for each subplot
    fig.colorbar(points, ax=ax, orientation='vertical')

# Adjust layout to avoid overlap
plt.tight_layout()

# Adding common axis labels
fig.text(0.5, 0.07, 'A2', ha='center', va='center')
fig.text(0.07, 0.5, 'eta2', ha='center', va='center', rotation='vertical')

# Adjust the top margin to make space for the title
plt.subplots_adjust(top=0.9)

# Adding a common title
fig.suptitle('GGL $\Delta \chi^2$ - $A_1$ and $\eta_1$ constant while $A_2$ and $\eta_2$ varies', fontsize=24)

plt.show()
```
